```python
import pandas as pd

uri = "../input/tracking.csv"
dados = pd.read_csv(uri)
dados.head()
```

```python
x= dados[["home","how_it_works","contact"]]
y= dados[["bought"]]
```

```python
mapa = { 
    "home" : "principal",
    "how_it_works" : "como_funciona",
    "contact" : "contato",
    "bought" : "comprou"
}
dados = dados.rename(columns = mapa)
```

```python
x= dados[["principal","como_funciona","contato"]]
y= dados[["comprou"]]

x.head()
#y.head()
```

```python
dados.shape
```

```python
treino_x = x[:75]
treino_x.shape
```

```python
treino_y = y[:75]
treino_y.shape
```

```python
teste_x = x[75:]
teste_y = y[75:]
teste_y.shape
print("Treinaremos com {} elementos e testaremos com {} elementos".format(len(treino_x), len(teste_x)))
```

```python
from sklearn.svm import LinearSVC #importa o modelo que queremos
from sklearn.metrics import accuracy_score #importa as métricas de acurácia

modelo = LinearSVC() #cria o modelo svc
modelo.fit(treino_x, treino_y) #manda treinar em cima do treino x e treino y
previsoes = modelo.predict(teste_x) #testa no teste x e teste y, prever

#compara as previsoes ou testes com o teste y, é a acurácia ou taxa de acertos
acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi {:.2f}%".format(acuracia))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import numpy as np

SEED = np.random.seed(20)#random.randrange(0, 101)

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                         random_state = SEED, test_size = 0.25,
                                                         stratify = y)
print("Treinaremos com {} elementos e testaremos com {} elementos".format(len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi de {:.2f}%".format(acuracia))
```
