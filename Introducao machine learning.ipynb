```python
# features (1 - sim, 0 - não)
# pêlo longo?
# perna curta?
# late au au?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

# 1 => porco, 0 => cachorro
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1,1,1,0,0,0] #labels / etiquetas
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(treino_x, treino_y)
```

```python
animal_misterioso = [1,1,1]
model.predict([animal_misterioso])
if(animal_misterioso == [1]):
    print("É um porco")
else:
    print("É um cachorro")
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

teste_x = [misterio1, misterio2, misterio3]
teste_y = [0, 1, 1]
```

```python
previsoes = model.predict(teste_x)
print (previsoes)
previsoes == teste_y
#if(previsoes.all == sequencia_correta_teste):
#if(previsoes in sequencia_correta_teste == True)
#    print("predição acertou")
#else:
#    print("predição errou")
#    print("A sequência correta era {}, mas foi passada a sequência {}".format(sequencia_correta_teste, previsoes))
```

```python
corretos = (previsoes == teste_y).sum()
total = len(teste_x)
taxa_de_acerto = corretos/total
print("Taxa de acerto: {:.2f}%".format(taxa_de_acerto *100))
```

```python

from sklearn.metrics import accuracy_score

taxa_de_acerto = accuracy_score(teste_y, previsoes)
print("Taxa de acerto: {:.2f}%".format(taxa_de_acerto * 100))
```
